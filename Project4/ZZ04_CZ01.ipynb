- Wykres 1: $N(0, 1)$, $N(0, 0.75)$, $N(0, 1.25)$, $N(-2, 1)$, $N(2, 1)$

```python
fig, ax = plt.subplots(2,3,figsize=(20,12))
fig.suptitle("Podstawowe rozkłady teoretyczne",fontsize=22)
x = np.linspace(-7.5,7.5,1000)
xlim12 = (-6,6)
ylim12 = (-0.01,0.6)

sigma = [0.75,1.25,1,1]
mean = [0,0,-2,2]
col = ["#EAC435","#E40066","#03CEA4","orchid"]  

ax[0,0].set_title("Rozkład normalny",fontsize=18)
ax[0,0].plot(x,sp.stats.norm.pdf(x,0,1),color="#345995",label="N(0,1)")
for i in range(len(sigma)):
    ax[0,0].plot(x,sp.stats.norm.pdf(x,mean[i],sigma[i]),color=col[i],label="N("+str(mean[i])+","+str(sigma[i])+")")
ax[0,0].legend(title="Rozkład")
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("pdf")
ax[0,0].set_xlim(xlim12)
ax[0,0].set_ylim(ylim12)

t = [1,2,5,30]
ax[0,1].set_title("Rozkład t-Studenta",fontsize=18)
ax[0,1].plot(x,sp.stats.norm.pdf(x,0,1),color="#345995",label="N(0,1)")
for i in range(len(t)):
    ax[0,1].plot(x,sp.stats.t.pdf(x,t[i]),color=col[i],label="t("+str(t[i])+")")
ax[0,1].legend(title="Rozkład")
ax[0,1].set_xlabel("x")
ax[0,1].set_ylabel("pdf")
ax[0,1].set_xlim(xlim12)
ax[0,1].set_ylim(ylim12)

x = np.linspace(0,15,1000)

chi2 = [1,2,3,5,8]
col.insert(0,"#345995")
ax[0,2].set_title("Rozkład chi-kwadrat",fontsize=18)
for i in range(len(chi2)):
    ax[0,2].plot(x,sp.stats.chi2.pdf(x,chi2[i]),color=col[i],label="chi2("+str(chi2[i])+")")
ax[0,2].legend(title="Rozkład")
ax[0,2].set_xlabel("x")
ax[0,2].set_ylabel("pdf")
ax[0,2].set_xlim((0,15))
ax[0,2].set_ylim((-0.01,0.6))

x = np.linspace(0,7.5,1000)
xlim456 = (-0.01,7.5)
ylim456 = (0,1.5)

dd = [1,2,5,10,20]
ax[1,0].set_title("Rozkład F Snedecora",fontsize=18)
for i in range(len(dd)):
    ax[1,0].plot(x,sp.stats.f.pdf(x,10,dd[i]),color=col[i],label="F(10,"+str(dd[i])+")")
ax[1,0].legend(title="Rozkład")
ax[1,0].set_xlabel("x")
ax[1,0].set_ylabel("pdf")
ax[1,0].set_xlim(xlim456)
ax[1,0].set_ylim(ylim456)

dn = [1,2,5,10,20]
ax[1,1].set_title("Rozkład F Snedecora",fontsize=18)
for i in range(len(dd)):
    ax[1,1].plot(x,sp.stats.f.pdf(x,dn[i],10),color=col[i],label="F("+str(dd[i])+",10)")
ax[1,1].legend(title="Rozkład")
ax[1,1].set_xlabel("x")
ax[1,1].set_ylabel("pdf")
ax[1,1].set_xlim(xlim456)
ax[1,1].set_ylim(ylim456)

d = [1,2,5,25,50]
ax[1,2].set_title("Rozkład F Snedecora",fontsize=18)
for i in range(len(dd)):
    ax[1,2].plot(x,sp.stats.f.pdf(x,d[i],d[i]),color=col[i],label="F("+str(d[i])+","+str(d[i])+")")
ax[1,2].legend(title="Rozkład")
ax[1,2].set_xlabel("x")
ax[1,2].set_ylabel("pdf")
ax[1,2].set_xlim(xlim456)
ax[1,2].set_ylim(ylim456)
```

#### b) Charakterystyka rozkładów
Na podstawie przygotowanych wykresów scharakteryzuje opisane rozkłady teoretyczne i odpowiedz na pytanie w jaki sposób zmiana wartości parametrów rozkładów wpływa na położenie i kształt ich funkcji gęstości prawdopodobieństwa.

***Rozkład normalny*** - jest określany przez dwa parametry: wartość oczekiwaną i odchylenie standardowe. Wartość oczekiwana wskazuje miejsce, w którym znajduje się środek rozkładu, natomiast wraz z modyfikacją odchylenia standardowego jego szerokość i wysokość się zmienia. Im niższe odchylenie standardowe tym rozkład jest wyższy i cieńszy. 

***Rozkład t-studenta*** parametrem określającym ten rozkład jest liczba stopni swobody. Dla 30 stopni swobody rozkład wygląda niemal identycznie jak N(0,1). Przy mniejszej liczbie stopni jest niższy, szerszy i bardziej spłaszczony niż rozkład normalny.

***Rozkład chi-kwadrat*** parametrem określającym ten rozkład jest liczba stopni swobody. Przy niskiej liczbie stopni swobody jest on prawostronnie asymetryczny. Wraz ze wzrostem parametru zaczyna być bardziej symetryczny i bardziej płaski. Przyjmuje tylko dodatnie wartości.

***Rozkład F Snedecora*** parametrami określającymi ten rozkład są stopnie swobody kolejno dla licznika i mianownika. Rozkład ten jest prawostronnie asymetryczny. Zwiększenie tylko jednego stopnia swobody sprawia, że wartość modalna przesuwa się w prawą stronę oraz wysokość rozkładu jest większa.
W przypadku wzrostu wartości obu stopni swobody rozkład zaczyna być coraz bardziej symetryczny. Przyjmuje tylko dodatnie wartości.

### Zadanie 2: Praca z rozkładami teoretycznymi [10 pkt]

#### a) Wyliczanie prawdopodobieństw na podstawie wartości

Oblicz i zapisz do zmiennych o podanych nazwach następujące wartości prawdopodobieństwa:
- `p1` - $P(X < -1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład t-Studenta o 10 stopniach swobody - $t(10)$,
- `p2` - $P(X > 2)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 50 stopniach swobody - $F(5, 50)$,
- `p3` - $P(3 < X < 10)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 4 stopniach swobody - $\chi^2(4)$,
- `p4` - $P(X < -2.75 | X > 1.5)$, gdzie $X$ to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.5 i odchyleniu standardowym równym 1.5 - $N(-0.5, 1.5)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wyznaczonej wartości prawdopodobieństwa. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość prawdopodobieństwa podaną do 4 miejsca po przecinku (np. P(X < -1.5) = x.xxxx, gdzie x.xxxx to obliczona wartość).

```python
# p1
p1 = sp.stats.t.cdf(-1.5,10)
```

```python
x_t = np.linspace(sp.stats.t.ppf(0.0001, 10), sp.stats.t.ppf(0.9999, 10), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_t, sp.stats.t.pdf(x_t, 10), 'black', lw=2)
plt.fill_between(x_t, sp.stats.t.pdf(x_t, 10), where=(x_t < -1.5), color='green', alpha=0.5)
plt.title(f'$P(X < -1.5) = {p1:.4f}$')
plt.xlim(-6,6)
plt.ylim(-0.01, 0.41)
plt.xlabel('x')
plt.ylabel('y')
```

```python
# p2
p2 = 1 - sp.stats.f.cdf(2,5,50)
```

```python
x_f = np.linspace(sp.stats.f.ppf(0.0001, 5, 50), sp.stats.f.ppf(0.9999, 5, 50), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_f, sp.stats.f.pdf(x_f, 5, 50), 'black', lw=2)
plt.fill_between(x_f, sp.stats.f.pdf(x_f, 5, 50), where=(x_f > 2), color='cornflowerblue', alpha=0.5)
plt.xlim(-0.3, 6.6)
plt.ylim(-0.01, 0.78)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'$P(X > 2) = {p2:.4f}$')
```

```python
# p3
p3 = sp.stats.chi2.cdf(10,4) - sp.stats.chi2.cdf(3,4)
```

```python
x_chi2 = np.linspace(sp.stats.chi2.ppf(0.0001, 4), sp.stats.chi2.ppf(0.9999, 4), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_chi2, sp.stats.chi2.pdf(x_chi2, 4), 'black', lw=2)
plt.fill_between(x_chi2, sp.stats.chi2.pdf(x_chi2, 4), where=((x_chi2 > 3) & (x_chi2 < 10)), color='indianred', alpha=0.5)
plt.xlim(-1, 24.5)
plt.ylim(-0.01, 0.19)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'$P(3 < X < 10) = {p3:.4f}$')
```

```python
# p4
p4 = sp.stats.norm.cdf(-2.75,-0.5,1.5) + 1 - sp.stats.norm.cdf(1.5,-0.5,1.5)
```

```python
x_norm = np.linspace(sp.stats.norm.ppf(0.0001, -0.5, 1.5), sp.stats.norm.ppf(0.9999,-0.5, 1.5), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_norm, sp.stats.norm.pdf(x_norm, -0.5, 1.5), 'black', lw=2)
plt.fill_between(x_norm, sp.stats.norm.pdf(x_norm, -0.5, 1.5),where=(x_norm < -2.75), color='thistle', alpha=0.5)
plt.fill_between(x_norm, sp.stats.norm.pdf(x_norm, -0.5, 1.5),where=(x_norm > 1.5), color='thistle', alpha=0.5)
plt.xlim(-6.5, 5.5)
plt.ylim(-0.01, 0.28)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'P(X < -2.75 | X > 1.5) = {p4:.4f}')
```

#### b) Wyliczanie wartości na podstawie prawdopodobieństw

Rozwiąż i zapisz do zmiennych o podanych nazwach rozwiązania następujących równań:
- `s1` - $P(X < s) = 0.25$, gdzie X to zmienna losowa ciągła posiadająca rozkład chi-kwadrat o 5 stopniach swobody - $\chi^2(5)$,
- `s2` - $P(X > s) = 0.15$, gdzie X to zmienna losowa ciągła posiadająca rozkład F Snedecora o 5 i 20 stopniach swobody - $F(5, 20)$,
- `s3` - $P(-1.25 < X < s) = 0.35$, gdzie x to zmienna losowa ciągła posiadająca rozkład normalny o wartości oczewiwanej równej -0.4 i odchyleniu standardowym równym 0.75 - $N(-0.4, 0.75)$,

Każde z rozwiązań zilustruj wykresem zawierającym krzywą gęstości prawdopodobieństwa rozkładu z zaznaczonym kolorem fragmentem pola pod krzywą odpowiadający wartości prawdopodobieństwa określonej przez rozwiązywane równanie. Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:
- ustaw zakres osi x od wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.0001 do wartości zmiennej $X$, dla której dystrybuanta analizowanego rozkładu przyjmuje wartość 0.9999,
- oznacz krawędź obszaru, na której znajduje się wyznaczona wartość $s$ linią w wyróżniającym się kolorze,
- w tytule wykresu zawrzyj uzupełniony wzór na prawdopodobieństwo uzupełniony o wyliczoną wartość zmiennej $X$ podaną do 4 miejsca po przecinku (np. P(X < s.ssss) = 0.25, gdzie s.ssss to obliczona wartość).

```python
# s1
s1 = sp.stats.chi2.ppf(0.25,5)
```

```python
x_chi2 = np.linspace(sp.stats.chi2.ppf(0.0001, 5), sp.stats.chi2.ppf(0.9999, 5), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_chi2, sp.stats.chi2.pdf(x_chi2, 5), 'black', lw=2)
plt.fill_between(x_chi2, sp.stats.chi2.pdf(x_chi2, 5),where=(x_chi2 < s1), color='green', alpha=0.5)

plt.vlines(x = s1, ymin = 0, ymax = 0.152, colors = 'red')
plt.xlim(-0.9, 27)
plt.ylim(-0.01, 0.16)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'P(X <  {s1:.4f}) = 0.25')
```

```python
# s2
s2 = sp.stats.f.isf(0.15,5,20)
```

```python
x_f= np.linspace(sp.stats.f.ppf(0.0001, 5, 20), sp.stats.f.ppf(0.9999, 5, 20), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_f, sp.stats.f.pdf(x_f, 5,20), 'black', lw=2)
plt.fill_between(x_f, sp.stats.f.pdf(x_f, 5, 20),where=(x_f > s2), color='cornflowerblue', alpha=0.5)

plt.vlines(x = s2, ymin = 0, ymax = 0.19, colors = 'blue')
plt.xlim(-0.3, 9.9)
plt.ylim(-0.01, 0.78)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'P(X > {s2:.4f}) = 0.15')
```

```python
# s3
s3 = sp.stats.norm.ppf(0.35 + sp.stats.norm.cdf(-1.25,-0.4,0.75), -0.4, 0.75)  
```

```python
x_norm = np.linspace(sp.stats.norm.ppf(0.0001, -0.4, 0.75), sp.stats.norm.ppf(0.9999, -0.4, 0.75), 1000)
plt.figure(figsize=(5, 5))
plt.plot(x_norm, sp.stats.norm.pdf(x_norm, -0.4, 0.75), 'black', lw=2)
plt.fill_between(x_norm, sp.stats.norm.pdf(x_norm, -0.4, 0.75), where=((x_norm > -1.25) & (x_norm < s3)), color='thistle', alpha=0.5)

plt.vlines(x = s3, ymin = 0, ymax = 0.529, colors = 'violet')
plt.vlines(x = -1.25, ymin = 0, ymax = 0.275, colors = 'violet')
plt.xlim(-3.5, 2.7)
plt.ylim(-0.01, 0.55)
plt.xlabel('x')
plt.ylabel('pdf')
plt.title(f'P(-1.25 < X < {s3:.4f}) = 0.35')
```
