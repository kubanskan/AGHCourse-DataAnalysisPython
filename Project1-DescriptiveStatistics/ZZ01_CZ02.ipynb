### Zadanie 3: Konstrukcja szeregów rodzielczych - wersja podstawowa [10 pkt]

#### a) Forma tabelaryczna
Przygotuj szeregi rozdzielcze VEI w formie tablic zawierających następujące elementy:

> Indeks: VEI </br>
> Wartości VEI posortowane od najmniejszej do największej.

> Kolumny: f, cf, rf, crf </br>
> Zliczenia poszczególnych kategorii (ang. frequency, absolute frequency), zliczenia skumulowane poszczególnych kategorii (ang. cumulative frequency, cumulative absolute frequency),  rf - częstość zliczeń poszczególnych kategorii (ang. relative frequency), crf - skumulowana częstość zliczeń poszczególnych kategorii (ang. cumulative relative frequency)

Przygotuj osobne tablice dla wszystkich erupcji (`vei_data_1`), erupcji potwierdzonych (`vei_data_2`) i erupcji niepotwierdzonych (`vei_data_3`).

Wyświetl przygotowane tablicę w taki sposób, aby liczby w kolumnach f i cf oraz indeks tablicy VEI wyświetlały się bez miejsc po przecinku, a liczby z kolumn rf i crf z dokładnością do 4 miejsca po przecinku.

```python
# Wszystkie erupcje
# YOUR CODE HERE
vei_data_1 = eruptions_data.value_counts(subset='VEI').to_frame().sort_index().rename(columns = {'count':'f'})
vei_data_1['cf'] = vei_data_1['f'].cumsum()
vei_data_1['rf'] = vei_data_1['f'] / vei_data_1['f'].sum()
vei_data_1['crf'] = vei_data_1['cf'] / vei_data_1['f'].sum()
```

```python
vei_data_1.style.format({'rf':'{:.4f}', 'crf':'{:.4f}' }).format_index(precision = 0)
```

```python
# Potwierdzone erupcje
# YOUR CODE HERE
vei_data_2 = eruptions_data[eruptions_data['Eruption Category'] == 'Confirmed'].value_counts(subset='VEI').to_frame().sort_index().rename(columns = {'count':'f'})
vei_data_2['cf'] = vei_data_2['f'].cumsum()
vei_data_2['rf'] = vei_data_2['f'] / vei_data_2['f'].sum()
vei_data_2['crf'] = vei_data_2['cf'] / vei_data_2['f'].sum()
```

```python
vei_data_2.style.format({'rf':'{:.4f}', 'crf':'{:.4f}' }).format_index(precision = 0)
```

```python
# Niepotwierdzone erupcje
# YOUR CODE HERE
vei_data_3 = eruptions_data[eruptions_data['Eruption Category'] == 'Uncertain'].value_counts(subset='VEI').to_frame().sort_index().rename(columns = {'count':'f'})
vei_data_3['cf'] = vei_data_3['f'].cumsum()
vei_data_3['rf'] = vei_data_3['f'] / vei_data_3['f'].sum()
vei_data_3['crf'] = vei_data_3['cf'] / vei_data_3['f'].sum()
```

```python
vei_data_3.style.format({'rf':'{:.4f}', 'crf':'{:.4f}' }).format_index(precision = 0)
```

```python
### Komórka testowa
# Test 1: Wszystkie erupcje
assert (vei_data_1.index.name == 'VEI') == True
assert len(set(list(vei_data_1.columns)) & set(['f', 'cf', 'rf', 'crf'])) == 4
assert list(vei_data_1['f'].index) == [0, 1, 2, 3, 4, 5, 6, 7]
assert list(vei_data_1['f'].values) == [1007, 1384, 3991, 1152, 511, 180, 53, 7]
# Test 2: Tylko potwierdzone erupcje
assert (vei_data_2.index.name == 'VEI') == True
assert len(set(list(vei_data_2.columns)) & set(['f', 'cf', 'rf', 'crf'])) == 4
assert list(vei_data_2['f'].index) == [0, 1, 2, 3, 4, 5, 6, 7]
assert list(vei_data_2['f'].values) == [803, 1227, 3727, 1133, 510, 180, 53, 7]
# Test 3: Tylko niepotwierdzone erupcje
assert (vei_data_3.index.name == 'VEI') == True
assert len(set(list(vei_data_3.columns)) & set(['f', 'cf', 'rf', 'crf'])) == 4
assert list(vei_data_3['f'].index) == [0, 1, 2, 3, 4]
assert list(vei_data_3['f'].values) == [204, 157, 264, 19, 1]
```

```python
# YOUR CODE HERE
fig, ax = plt.subplots(1,3, figsize = (25,10))
colorf = 'tab:cyan'
colorcf = 'tab:red'

ax1 = ax[0].twinx()
ax[0].bar(vei_data_1.index, vei_data_1['f'], color = colorf) 
ax1.plot(vei_data_1.index, vei_data_1['cf'], color = colorcf, lw = 2)
ax[0].set_xlabel('VEI', fontsize = 18)
ax[0].set_ylabel('F', color = colorf, fontsize = 18)
ax1.set_ylabel('CF', color = colorcf, fontsize = 18)
ax1.tick_params(labelcolor = colorcf, labelsize = 15)
ax[0].tick_params(axis = 'y', labelcolor = colorf)
ax[0].tick_params(labelsize=15)
ax[0].set_title('Wszystkie erupcje',  fontsize = 22)
ax[0].set_ylim(0,4200)
ax[0].set_xlim(0,7)
ax1.set_ylim(0,8500)

ax2 = ax[1].twinx()
ax[1].bar(vei_data_2.index, vei_data_2['f'], color = colorf) 
ax2.plot(vei_data_2.index, vei_data_2['cf'], color = colorcf, lw = 2)
ax2.tick_params(axis = 'y', labelcolor = colorcf, labelsize = 15)
ax[1].set_xlabel('VEI', fontsize = 18)
ax[1].set_ylabel('F', color = colorf, fontsize = 18)
ax2.set_ylabel('CF', color = colorcf, fontsize = 18)
ax[1].tick_params(axis = 'y', labelcolor = colorf)
ax[1].tick_params(labelsize=15)
ax[1].set_title('Potwierdzone erupcje', fontsize = 22)
ax[1].set_ylim(0,4200)
ax[1].set_xlim(0,7)
ax2.set_ylim(0,8500)

# Add extra values for vei_data_3 to have exual axes
extended_vei_data = vei_data_3.copy()
last_y_value = extended_vei_data['cf'].iloc[-1]
for x in range(5, 8): 
    extended_vei_data.loc[x, 'cf'] = last_y_value

ax3 = ax[2].twinx()
ax[2].bar(vei_data_3.index, vei_data_3['f'], color = colorf) 
ax3.plot(extended_vei_data.index, extended_vei_data['cf'], color = colorcf, lw = 2)
ax[2].set_xlabel('VEI', fontsize = 18)
ax[2].set_ylabel('F', color = colorf, fontsize = 18)
ax3.set_ylabel('CF', color = colorcf, fontsize = 18)
ax3.tick_params(labelcolor = colorcf, labelsize=15)
ax[2].tick_params(axis = 'y', labelcolor = colorf)
ax[2].tick_params(labelsize=15)
ax[2].set_title('Niepotwierdzone erupcje', fontsize = 22)
ax[2].set_ylim(0,4200)
ax[2].set_xlim(0,7)
ax3.set_ylim(0,8500)


fig.suptitle('Ilość wystąpień erupcji', fontsize = 25)
fig.tight_layout(pad = 1.3)
```
