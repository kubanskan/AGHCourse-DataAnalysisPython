#### a) Całość danych
Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

```python
# YOUR CODE HERE
pd.option_context('mode.use_inf_as_na', True)
def plotHist(row, col, title, color, year,  maska):
    sns.histplot(recession_data_1[maska][year], ax = ax[row,col], fill = False, color = color, element = 'poly', stat = 'probability', bins = np.arange(-20, 35, 5), label = year)
    ax[row, col].set_title(title)
    ax[row, col].set_xlabel('GPD (%)')
    ax[row, col].set_ylabel('Częstość')
    ax[row, col].legend(title='Year')
    ax[row, col].yaxis.grid(True)
    ax[row, col].set_xlim(-20,35)
    ax[row, col].set_ylim(0, 0.8)

    
fig, ax = plt.subplots(2,5,figsize = (20,11))
years = recession_data_1.columns[3:]
colors = ['green', 'pink', 'orange', 'red', 'blue', 'yellow']

color_dict = {years[i]: colors[i] for i in range(len(years))}

for year in years:
    
    col = 0
    color = color_dict[year] 
    plotHist(0, 0, f'All groups', color, year, [True] * len(recession_data_1))
    
    for group in recession_data_1['Income Group'].unique():
        col += 1
        plotHist(0,col, f'Income Group: {group}', color, year, recession_data_1['Income Group'] == group)

    for group in recession_data_1['Region Name'].unique():   
        col -= 1
        plotHist(1, col, f'Region Name: {group}', color, year, recession_data_1['Region Name'] == group)
        
fig.suptitle('Diagramy częstości', fontsize = 20)
```

#### c) Dyskusja wyników

Zastanów się jakie są mocne i słabe strony poszczególnych typów wykresów wykonanych w tym zadaniu. W jakich przypadkach wykorzystanie konkretnych typów wykresów do wizualizacji może sprawiać problemy?

Diagramy częstości dobrze demonstrują podobieństwa pomiędzy krzywymi, ogólną tendencję oraz ich kształt. Nie da się jednak porównać na jednym wykresie np. różnych stopni zamożności, trzeba dla każdej kategorii wykonać osobny wykres, ponieważ kolor już odróżnia nam konkretne lata. Problematyczne może też się okazać określenie konkretnych wartości i czytelność, ponieważ linie czasem nachodzą na siebie i się rozmywają. Ten typ wykresów lepiej sprawdza się dla niezbyt dużej ilości krzywych na jednym wykresie, oraz najlepiej żeby były one lekko zróżnicowane.

Zaletą dalszych wykresów jest to, że na jednym wykresie możemy porównać państwa o różnych stopniach zamożności oraz różne regiony.

Boxplot w sposób prosty i intuicyjny pokazuje ogólny rozkład danych - medianę oraz kwartyle. Można też dzięki nim wykryć wartości odstające. W naszym przypadku dobrze przedstawia jaki stosunek mają do siebie dane. Mimo tego, że wysokość pudełka świadczy o rozproszeniu danych to jednak ten typ wykresów nie demonstruje dokładnej postaci rozkładu danych. 

Wykresy skrzypcowe pokazują gęstość prawdopodobieństwa. Np. dla roku 2006 i 2007 widzimy subtelną różnicę pomiędzy danymi, podczas gdy dla wykresu pudełkowego ciężko było dostrzec większe różnice. Niestety dla danych z podziałem na regiony wykres staje się nieco cięższy do interpretacji przez dużą ilość rozkładów. Tu już czytelniejszy wydaje się być boxplot, chyba że chce się wychwycić pewne różnice między rozkładami.

Swarmplot pokazuje nam całość danych, co jest z jednej strony korzystne bo nie tracimy żadnych informacji. Możemy dostrzec zagęszczenia punktów w niektórych miejscach. Dla podziału na regiony lub stopnie zamożności można dostrzec ciekawe zjawiska w danych, jak np. Europa w 2009, gdzie możemy dostrzec grupę 4 państw wykazujących się dużą recesją gospodarczą, od głównej grupy odstają też widocznie 2 inne państwa z dużym spadkiem. Nie pokazują jednak rozkładu danych, tak jak wcześniejsze wykresy, co dla dużej ilości danych może być minusem.

Stripplot podobnie jak powyższy typ pokazuje wszystkie dane. Z jednej strony jest prostrzy do interpretacji, pozwala łatwiej dostrzec zagęszczenia punktów, ale z drugiej prowadzi do nakładania się na siebie punktów. Dla dużej ilości danych, w naszym przypadku dla całości, wykres dostarcza mało informacji. (lepiej sprawdza się swarmplot, który zapobiega nakładaniu się punktów)
