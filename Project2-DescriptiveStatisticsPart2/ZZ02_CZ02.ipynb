Wyświetl przygotowane tablice w taki sposób aby zawarte w tablicy liczby (za wyjątkiem lat i granic przedziałów w nagłówkach kolumn i wierszy) wyświetlały się z dokładnością do 2 miejsca po przecinku.

#### a) Całość danych
Pierwsza tablica (`recession_table_1`) powinna zawierać szeregi skonstruowane w obarciu o całość danych.

 > Indeks: Year </br>
 > Wszystkie lata, dla których zostały wyznaczone wartości parametru GDP growth (annual %)
 
 > Kolumny: (-20, -15],	(-15, -10], ... , (30, 35] </br>
 > Kolejne przedziały wartości zdefiniowane jako `pandas.IntervalIndex`

```python
# YOUR CODE HERE
idx = pd.IntervalIndex.from_breaks([*range(-20,36,5)])
df = pd.melt(recession_data_1, value_vars=['2006','2007','2008','2009','2010','2011'], var_name='Year', value_name='GDP' )
recession_table_1 = df.groupby(['Year',  pd.cut(df['GDP'],idx)], observed=False).count()
recession_table_1.columns = ['count']
recession_table_1 = recession_table_1.reset_index().set_index(['Year']).pivot(columns='GDP',values='count')
```

```python
recession_table_1 = recession_table_1 / len(recession_data_1)
recession_table_1.style.format(precision = 2)
```

```python
### Komórka testowa
# Kolumny
lower_boundaries = np.arange(-20, 35, 5)
for i in range(len(lower_boundaries)):
    assert recession_table_1.columns[i] == pd.Interval(lower_boundaries[i], lower_boundaries[i]+5, closed='right')
# Indeks
assert recession_table_1.index.name == "Year"
assert len(set(list(recession_table_1.index)) & set([2006, 2007, 2008, 2009, 2010, 2011])) == 6 \
    or len(set(list(recession_table_1.index)) & set(['2006', '2007', '2008', '2009', '2010', '2011'])) == 6
# Wartości w obrębie tablicy
assert recession_table_1.values.min() >= 0.0
assert recession_table_1.values.max() <= 1.0
assert np.all(np.isclose(recession_table_1.sum(axis=1), 1.0))==True
```

#### b) Podział na regiony geograficzne

Zestaw wyniki na pojedynczej figurze (siatka wykresów 2x5). W górnym wierszu zawrzyj diagram skonstruowany w oparciu o całość danych oraz cztery diagramy skonstruowane indywidualnie dla poszczególnych stopni zamożności, a w dolnym wierszu pięć dziagramów skonstruowanych indywidualnie dla poszczególnych rejonów geograficznych). Przygotuj figurę zgodnie z wytycznymi z nagłówka notatnika. Dodatkowo:

Uwaga! Zadanie należy wykonać w oparciu o wczytane do notatnika szeregi szczegółowe, a nie w oparciu o szeregi rozdzielcze częstości przygotowane w zadaniu 2.

```python
# YOUR CODE HERE
def plotHist(row, col, title, color, year,  maska):
    sns.histplot(recession_data_1[maska][year], ax = ax[row,col], fill = False, color = color, element = 'poly', stat = 'probability', bins = np.arange(-20, 35, 5), label = year)
    ax[row, col].set_title(title)
    ax[row, col].set_xlabel('GPD (%)')
    ax[row, col].set_ylabel('Częstość')
    ax[row, col].legend(title='Year')
    ax[row, col].yaxis.grid(True)
    ax[row, col].set_xlim(-20,35)
    ax[row, col].set_ylim(0, 0.8)

    
fig, ax = plt.subplots(2,5,figsize = (24,12))
years = recession_data_1.columns[3:]
colors = ['green', 'pink', 'orange', 'red', 'blue', 'yellow']

color_dict = {years[i]: colors[i] for i in range(len(years))}

for year in years:
    
    col = 0
    color = color_dict[year] 
    plotHist(0, 0, f'All groups', color, year, [True] * len(recession_data_1))
    
    for group in recession_data_1['Income Group'].unique():
        col += 1
        plotHist(0,col, f'Income Group: {group}', color, year, recession_data_1['Income Group'] == group)

    for group in recession_data_1['Region Name'].unique():   
        col -= 1
        plotHist(1, col, f'Region Name: {group}', color, year, recession_data_1['Region Name'] == group)
        
fig.suptitle('Diagramy częstości', fontsize = 20)
```

```python
# Boxplot
#melt funkcja do zlaczenia tablic, hue
data_1 = pd.melt(recession_data_1.iloc[:, 3:], var_name='Year', value_name='GDP')
data_2 = pd.melt(recession_data_1, value_vars=recession_data_1.iloc[:, 3:], id_vars="Income Group", var_name='Year', value_name='GDP')
data_3 = pd.melt(recession_data_1, value_vars=recession_data_1.iloc[:, 3:], id_vars="Region Name", var_name='Year', value_name='GDP')

fig, ax = plt.subplots(3,1,figsize = (12,17))
fig.suptitle("Boxplots", fontsize=18)

sns.boxplot(data=data_1, x='Year', y='GDP',  ax=ax[0], color='yellow', linewidth=1, showmeans=True, meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black"}, flierprops={"marker": "o", "markeredgewidth":1.5, "markerfacecolor":'none'})
ax[0].yaxis.grid(True)
ax[0].axhline(y=0, color="red")
ax[0].set_ylim([-30,40])
ax[0].set_title('Whole Data')

sns.boxplot(data=data_2, x='Year', y='GDP',  ax=ax[1], hue="Income Group", palette='coolwarm', showmeans=True, meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black"}, flierprops={"marker": "o", "markeredgewidth":1.5, "markerfacecolor":'none'})
ax[1].yaxis.grid(True)
ax[1].axhline(y=0, color="red")
ax[1].set_ylim([-30,40])
ax[1].set_title('Grouped by Income Group')
ax[1].legend(title="Income Group", ncols=4)

color_palette=["#F9C04F", "#5FA1BD", "#565554", "#70D26D", "#D8453D"]
sns.boxplot(data=data_3, x='Year', y='GDP',  ax=ax[2], hue="Region Name", palette=color_palette, showmeans=True, meanprops={"marker":"o","markerfacecolor":"white","markeredgecolor":"black"}, flierprops={"marker": "o", "markeredgewidth":1.5, "markerfacecolor":'none'})
ax[2].yaxis.grid(True)
ax[2].axhline(y=0, color="red")
ax[2].set_ylim([-30,40])
ax[2].set_title('Grouped by Region Name')
ax[2].legend(title="Region Name", ncols=5)
```
